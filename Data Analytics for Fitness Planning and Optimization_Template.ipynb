# **Data Cleaning and Data Pre-processing**

```python
# Clean and Trim the column values in the dataset.
To_clean_col=df.columns # it will fetch all the column name

for col in To_clean_col:
  if df[col].dtype=='object': ## Check if column is object/string type before cleaning
    df[col]=df[col].str.strip() #removes leading and trailing whitespaces
    df[col].str.title() # All the text values are captitalized

print("Clean and trim sucessfully done")
```

```python
# checking the null value is present in the datasets
print("Count of null values in each column are: ", df.isnull().sum(), sep='\n')
```

```python
# Remove duplicates from the dataset

#df.drop_duplicates(inplace=True) # delete the duplicate values from the dataset
#print("Shape of the DataFrame after removing duplicates:", df.shape)
if df.duplicated().sum()==0:
  print("No Duplicte values found in the dataset",df.shape)
else:
  print("Number of dupicates present in the dataset are : ", df.duplicated().sum())
  print("Removed all duplicate values in the dataset", df.drop_duplicates(inplace=True))
```

```python

# Rename the column names
df=df.rename(columns={"ID":"Cust_ID","Sex":"Gender","Level":"Weight Category"})
df.columns # fetch only the column name
```

```python
# Summary statistics
print("Summary Statics for dataset")
print(df.describe())
```

```python
print("Outlier Detection and Treatment Using IQR-Based method")

# Select only numerical columns for outlier treatment
numerical_cols = df.select_dtypes(include=np.number).columns

for column in numerical_cols:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Find outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    num_outliers = len(outliers)
    print(f"\nColumn '{column}': Found {num_outliers} outliers.")

    # Cap the outliers
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

    print(f"  Outliers in column '{column}' have been capped.")

print("\nOutlier treatment complete for all numerical columns.")
```

```python
import numpy as np

print("Re-checking for Outliers Using IQR-Based method on current DataFrame")

# Select only numerical columns for outlier detection
numerical_cols = df.select_dtypes(include=np.number).columns

outliers_found_after_capping = False

for column in numerical_cols:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Find outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    num_outliers = len(outliers)

    if num_outliers > 0:
        print(f"\nColumn '{column}': Found {num_outliers} outliers.")
        outliers_found_after_capping = True
    else:
        print(f"\nColumn '{column}': No outliers found by IQR method.")

if not outliers_found_after_capping:
    print("\nNo outliers detected in any numerical column by IQR method after previous treatment.")
```

```python

# Check skewness and apply transformations

numerical_cols = df.select_dtypes(include=np.number).columns

print("--- Skewness Check and Transformation ---")
print("Threshold for log transformation: Skewness > 0.5 or < -0.5")

for col in numerical_cols:
    initial_skewness = df[col].skew()
    print(f"\nColumn '{col}':")
    print(f"  Initial Skewness = {initial_skewness:.4f}")


print("\n--- Skewness check and transformation complete. ---")
```

```python
# Identify object columns that are suitable for categorical conversion
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']

print(f"Converting {len(categorical_cols)} columns from 'object' to 'category':")
for col in categorical_cols:
    df[col] = df[col].astype('category')
    print(f"  - '{col}' converted.")

print("\nData types after conversion:")
df.info()
```

```python
# Identify all object columns
object_cols = df.select_dtypes(include=['object']).columns
# Convert all identified columns to categorical
df[object_cols] = df[object_cols].astype('category')

print(df.dtypes)
```

# **EDA and Visualizations**

## Data Visualization: Countplots for Categorical Variables

A countplot is a type of bar chart that shows the counts of observations in Weight Category  of a categorical variable. It's useful for understanding the distribution of Weight Level data and identifying imbalances.

```python
#df=sns.load_dataset(df)
sns.countplot(data=df,x='Weight Category')
plt.title('Distribution of Weight Level')
plt.xlabel('Weight Level')
plt.ylabel('count')
plt.show()
```

## Data Visualization: Histogram for Numerical Variables

A histogram is a graphical representation that organizes a group of data points into user-specified ranges. It is useful for visualizing the distribution of numerical data, showing the frequency of values within different bins.

```python
# Create a histogram for the 'Age' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Data Visualization: Boxplot

A boxplot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It can also indicate outliers. Boxplots are particularly useful for comparing distributions across different groups.

```python
# Create a boxplot for 'Weight' grouped by 'Weight Category'
plt.figure(figsize=(12, 7))
sns.boxplot(x='Weight Category', y='Weight', data=df)
plt.title('Weight Distribution by Weight Category')
plt.xlabel('Weight Category')
plt.ylabel('Weight (kg)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Create a boxplot for 'Weight' grouped by 'Weight Category' and 'Gender'
plt.figure(figsize=(14, 8))
sns.boxplot(x='Weight Category', y='Weight', hue='Gender', data=df, palette='viridis')
plt.title('Weight Distribution by Weight Category and Gender')
plt.xlabel('Weight Category')
plt.ylabel('Weight (kg)')
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Data Visualization: Scatterplot

A scatterplot is a type of plot that displays values for typically two variables for a set of data. Each point on the scatterplot represents an observation in the dataset, and its position is determined by the values of the two variables. Scatterplots are useful for identifying relationships, trends, and clusters between variables.

```python
# Create a scatterplot for 'Height' vs 'Weight'
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Age', y='BMI', data=df, hue='Gender')
plt.title('Scatterplot of Age vs. BMI')
plt.xlabel('Age')
plt.ylabel('BMI)')
plt.grid(True, linestyle='--')
plt.show()
```
